```python
# importieren von matplotlib
import matplotlib.pyplot as plt
```

```python
# ein erster einfacher Plot
plt.plot([1, 5, 9, 12, 5, 3, 8])
plt.show()
```

```python
# nun der gleiche Plot mit anderer Darstellung
plt.plot([1, 5, 9, 12, 5, 3, 8],"o")
plt.show()
```

```python
# jetzt doch lieber mit einer gestrichelten Linie
plt.plot([1, 5, 9, 12, 5, 3, 8], "--")
plt.show()
```

```python
# und nun lieber in der Farbe magenta
plt.plot([1, 5, 9, 12, 5, 3, 8], "--m")
plt.show()
```

```python
# nun mal zwei Datenreihen (x,y) plotten
monatstage = list(range(1,32))
# print(monatstage)
messwerte = [2,4,3,4,5,3,2,6,3,6,4,2,6,3,7,4,6,7,2,1,4,5,0,4,6,7,6,6,7,3,4]
plt.plot(monatstage, messwerte, "--m")
plt.show()
```

```python
# nun spendieren wir mal eine Achsenbeschriftung
# nun mal zwei Datenreihen (x,y) plotten
monatstage = list(range(1,32))
# print(monatstage)
messwerte = [2,4,3,4,5,3,2,6,3,6,4,2,6,3,7,4,6,7,2,1,4,5,0,4,6,7,6,6,7,3,4]
plt.plot(monatstage, messwerte, "--m")
plt.xlabel("Tag des Monats")
plt.ylabel("Sonnenstunden")
plt.show()
```

```python
# man kann auch mehrer Datenreihen in einem Diagramm darstellen
# nun spendieren wir mal eine Achsenbeschriftung
# nun mal zwei Datenreihen (x,y) plotten
monate = list(range(1,13))
regentage = [2,4,3,4,5,3,2,6,3,6,4,2]
niederschlagsmenge = [220,124,210,45,330,58,190,257,209,113,227,97]
plt.plot(monate, regentage, "--m", monate, niederschlagsmenge, "--b")
plt.xlabel("Tag des Monats")
plt.ylabel("Niederschlag Tage/Menge")
plt.show()
```

```python
# eine weitere MÃ¶glichkeit der Formatierung 
#... viele Wege fÃ¼hren nach Rom
monate = list(range(1,13))
regentage = [2,4,3,4,5,3,2,6,3,6,4,2]
niederschlagsmenge = [220,124,210,45,330,58,190,257,209,113,227,97]
plt.plot(monate, regentage, color = "green", linewidth = 4.5, linestyle = "--")
plt.plot(monate, niederschlagsmenge, color = "blue", linewidth = 2, linestyle = "-")
plt.xlabel("Tag des Monats")
plt.ylabel("Niederschlag Tage/Menge")
plt.show()
```

```python
# FlÃ¤chen unter der Kurve einfÃ¤rben
monate = list(range(1,13))
regentage = [2,4,3,4,5,3,2,6,3,6,4,2]
niederschlagsmenge = [220,124,210,45,330,58,190,257,209,113,227,97]
plt.plot(monate, regentage, color = "green", linewidth = 4.5, linestyle = "--")
plt.plot(monate, niederschlagsmenge, color = "blue", linewidth = 2, linestyle = "-")
plt.fill_between(monate, 0, niederschlagsmenge, color='blue', alpha=.4)
plt.xlabel("Tag des Monats")
plt.ylabel("Niederschlag Tage/Menge")
plt.show()
```

```python
# ein Diagrammtitel fehlt noch
monate = list(range(1,13))
regentage = [2,4,3,4,5,3,2,6,3,6,4,2]
niederschlagsmenge = [220,124,210,45,330,58,190,257,209,113,227,97]
plt.plot(monate, regentage, color = "green", linewidth = 4.5, linestyle = "--")
plt.plot(monate, niederschlagsmenge, color = "blue", linewidth = 2, linestyle = "-")
plt.fill_between(monate, 0, niederschlagsmenge, color='blue', alpha=.1)
plt.xlabel("Tag des Monats")
plt.ylabel("Niederschlag Tage/Menge")
plt.title("mein Diagramm")
plt.show()
```

```python
# nun stellen wir die AbstÃ¤nde der "Ticks" auf der X-Achse ein
import numpy as np
monatstage = list(range(1,32))
# print(monatstage)
messwerte = [2,4,3,4,5,3,2,6,3,6,4,2,6,3,7,4,6,7,2,1,4,5,0,4,6,7,6,6,7,3,4]
plt.plot(monatstage, messwerte, "--m")
plt.xlabel("Tag des Monats")
plt.ylabel("Sonnenstunden")
plt.xticks(np.linspace(0, 30, 6,endpoint=True))
plt.show()
```

```python
# Arbeiten mit seaborn
# seaborn ist eine Bibliothek fÃ¼r das erstellen von Grafiken fÃ¼r statistische Auswerungen
# viele Beispiele und ausfÃ¼hliche Informationen zu seaborn findet man unter:
# https://seaborn.pydata.org/introduction.html
# Die folgenden Code-Beispiele wurden von https://seaborn.pydata.org/introduction.html Ã¼bernommen.
import seaborn as sns

# Apply the default theme
sns.set_theme()

# Load an example dataset
tips = sns.load_dataset("tips")

# Create a visualization
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size",
)
```

```python
# Ein weiteres Beispiel aus der Bibliothek seaborn zeigt, dass wenige Zeilen Code notwendig sind, 
# um kompliziert wirkende plots zu erstellen.
# Die folgenden Code-Beispiele wurden von https://seaborn.pydata.org/introduction.html Ã¼bernommen.

dots = sns.load_dataset("dots")
sns.relplot(
    data=dots, kind="line",
    x="time", y="firing_rate", col="align",
    hue="choice", size="coherence", style="choice",
    facet_kws=dict(sharex=False),
)
```

```python
# Ein weiteres Beispiel aus der Bibliothek seaborn zeigt, dass wenige Zeilen Code notwendig sind, 
# um kompliziert wirkende plots zu erstellen.
# Die folgenden Code-Beispiele wurden von https://seaborn.pydata.org/introduction.html Ã¼bernommen.

penguins = sns.load_dataset("penguins")
sns.jointplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", hue="species")
```
